```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Apply standard scaling to the selected features
df[['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']] = scaler.fit_transform(df[['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']])

# Verify that the scaling has been applied correctly
print(df.head())
```

```python
## Independent and dependent features
X = df.drop(labels=['Payment_default'],axis=1)
Y = df[['Payment_default']]
```

```python
from sklearn.impute import SimpleImputer ## HAndling Missing Values
from sklearn.preprocessing import StandardScaler # HAndling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding
## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
## Train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
## Model Training

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score 
```

```python
regression=LogisticRegression()
regression.fit(X_train,y_train)
```

```python
## Accuracy on train data
train_data_pred = regression.predict(X_train)
accuracy_on_train = accuracy_score(train_data_pred, y_train)
```

```python
print('Accuracy on train data:', accuracy_on_train)
```

```python
## Accuracy on test data
test_data_pred = regression.predict(X_test)
accuracy_on_test = accuracy_score(test_data_pred, y_test)
```

```python
print('Accuracy on test data:', accuracy_on_test)
```
